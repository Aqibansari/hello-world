```
import numpy as np
import torch
import torchvision
import matplotlib.pyplot as plt
```

```
mnist_dev = torchvision.datasets.MNIST('./data', train=True, download=True)
mnist_test = torchvision.datasets.MNIST('./data', train=True, download=True)
```

```
print(mnist_dev[0])
```

```
x_dev = mnist_dev.data / 255.0
y_dev = mnist_dev.targets
```

```
plt.imshow(x_dev[0])
```

```
num_train = 5000
x_train, y_train = x_dev[:num_train].flatten(1), y_dev[:num_train]
x_val, y_val = x_dev[num_train:].flatten(1), y_dev[num_train:]
```

```
num_features = 28*28
num_classes = 10

weight = torch.empty(num_features, num_classes).uniform_(-1, 1) * np.sqrt(2 / (num_features + num_classes))
weight.requires_grad_()
bias = torch.zeros(num_classes, requires_grad=True)
```

```
def model(input):
    """
    Args: 
        input: Tensor container features of shape (num_samples, num_features)
    Returns:
        logits: Tensor of logits, shape (num_samples, num_classes)
    """
    return log_softmax(input@weight + bias)

def log_softmax(input):
    return (input - input.logsumexp(-1, keepdims=True))\

def nll_loss(logits, target):
    return -logits[range(target.shape[0]), target].mean()

loss_fn = nll_loss

def get_accuracy(logits, target):
    pred = logits.argmax(dim=-1)
    return (pred == target).float().mean()
```

```
batch_size = 64
print(get_accuracy(model(x_train[:batch_size]), y_train[:batch_size]))
```

```
learning_rate = 0.6
num_epochs = 2

for epoch in range(num_epochs):
    for i in range(int(np.ceil(num_train / batch_size))):
        start_i = i * batch_size
        end_i = min(start_i * batch_size, num_train)
        x_batch_size = x_train[start_i:end_i]
        y_batch_size = y_train[start_i:end_i]

        logits = model(x_batch_size)

        loss = loss_fn(logits, y_batch_size)

        loss.backward()

        with torch.no_grad():
            weight -= learning_rate * weight.grad
            bias -= learning_rate * bias.grad
        weight.grad.zero_()
        bias.grad.zero_()
```

```
print(get_accuracy(model(x_train[:batch_size]), y_train[:batch_size]))
```
